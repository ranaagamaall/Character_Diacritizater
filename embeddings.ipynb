```python
from transformers import AutoTokenizer, AutoModel
import torch
import pandas as pd

tokenizer = AutoTokenizer.from_pretrained('CAMeL-Lab/bert-base-arabic-camelbert-ca')
model = AutoModel.from_pretrained('CAMeL-Lab/bert-base-arabic-camelbert-ca')
# # text = "قَوْلُهُ وَلَا تُكْرَهُ ضِيَافَتُهُ"
# text = ["مرحبا يا عالم.", "أهلا بكم في العالم."]
# encoded_input = tokenizer(text, return_tensors='pt', padding=True)
# output = model(**encoded_input)
```

```python
import pandas as pd
vocab = pd.read_csv("processed/vocab_train.csv", header=None)
vocab.head()
```

```python
tokenized_ids = tokenizer(list(vocab[0]), return_tensors="pt", padding=True)

with torch.no_grad():
    output = model(**tokenized_ids)
    embeddings = output.last_hidden_state

word_embeddings = []
for i in range(len(vocab)):
    word_embeddings.append(embeddings[i][1])


vocab["embedding"] = word_embeddings
# def get_word_embedding(word):
#     tokenizer = AutoTokenizer.from_pretrained('CAMeL-Lab/bert-base-arabic-camelbert-ca')
#     model = AutoModel.from_pretrained('CAMeL-Lab/bert-base-arabic-camelbert-ca')
#     encoded_input = tokenizer(word, return_tensors='pt', padding=True)
#     output = model(**encoded_input)
#     return output.last_hidden_state[0][1]

# vocab["embedding"] = vocab[0].apply(get_word_embedding)
```
